```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a great home for the long term",
    candidate_labels=["investment", "property", "comfort"],
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("Don't forget to use an <mask> so you don't get caught in the rain", top_k=2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Pramita, and I am a student majoring in Informatics Engineering.")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Python adalah bahasa tingkat tinggi yang mendukung berbagai paradigma pemrograman berorientasi objek, fungsional, dan imperatif."
question = "Apa itu Bahasa Pemrograman Python?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Bill Gates adalah seorang pengusaha, programmer, dan filantropis asal Amerika Serikat dikenal sebagai pendiri Microsoft Corporation, salah satu perusahaan perangkat lunak terbesar di dunia. Ia lahir pada 28 Oktober 1955 di Seattle, Washington."
question = "Pendiri Microsoft adalah?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

## HASIL ANALISIS

**Zero-short classification :**

Zero-shot classification adalah teknik dalam pemrosesan Natural Language Processing yang memungkinkan model untuk mengklasifikasikan teks ke dalam kategori tertentu tanpa pelatihan khusus pada kategori tersebut sebelumnya.

Contoh 1 : kalimat "This is a great home for the long term" lebih terkait dengan konsep "property" dibandingkan dengan label lain seperti "investment" atau "comfort." Skor yang tinggi menunjukkan bahwa model percaya kalimat tersebut lebih mencerminkan makna atau konteks dari label "property."

Contoh 2 : kalimat "This drink is good for body health" lebih terkait dengan konsep "Health" dibandingkan dengan label lain. Skor yang tinggi menunjukkan bahwa model percaya kalimat tersebut lebih mencerminkan makna atau konteks dari label tersebut.

Contoh 3 : kalimat "The boutique sells various models of clothing" lebih terkait dengan konsep "business" dibandingkan dengan label lain.

**Text-Generation**

Text generation adalah teknik dalam pemrosesan bahasa alami yang memungkinkan model menghasilkan teks secara otomatis berdasarkan prompt atau masukan tertentu

"text-generation" dengan model "distilgpt2" berfungsi untuk menghasilkan teks lanjutan berdasarkan kalimat awal yang diberikan. Dalam contoh ini, model dapat menciptakan kelanjutan yang relevan tentang apa yang mungkin diajarkan oleh Mr. Arifian dalam kursus Python untuk Data Science. Hasil ini menunjukkan kemampuan model dalam memahami konteks dan menyusun informasi secara alami.

**Fill-mask**

Teknik dalam pemrosesan bahasa alami yang memungkinkan model untuk mengisi kata yang hilang atau kosong dalam sebuah kalimat dengan kata yang paling cocok berdasarkan konteks

"fill-mask" mampu mengidentifikasi dan mengisi kata yang hilang dalam kalimat berdasarkan konteks yang diberikan. Contohnya kalimat diatas, model akan memberikan pilihan kata yang relevan seperti "umbrella" untuk melindungi dari hujan.

**Ner**

Pemrosesan bahasa alami yang digunakan untuk mengidentifikasi dan mengkategorikan entitas bernama dalam teks, seperti nama orang, organisasi, lokasi, tanggal, angka, dan lainnya

"ner" berhasil mengidentifikasi dan mengelompokkan entitas dalam kalimat "My name is Pramita, and I am a student majoring in Informatics Engineering". Model mengenali "Pramita" sebagai nama orang dan "Informatics Engineering" sebagai bidang studi. Kemampuan model dalam mengekstrak informasi penting dan mengkategorikannya secara akurat, yang berguna untuk analisis teks dan pemrosesan informasi.

**Question answering**

"question-answering" berhasil mengidentifikasi jawaban yang tepat untuk pertanyaan "Pendiri Microsoft adalah?" berdasarkan kalimat yang diberikan. Model berhasil mendeteksi informasi dari teks dan memberikan jawaban yang benar yaitu "Bill Gates." Ini menunjukan model dalam memahami dan mengekstraksi informasi yang dibutuhkan dari teks.

**Sentiment analysis**

Pipeline "sentiment-analysis" berhasil mengidentifikasi sentimen positif dalam kalimat "This car is very nice! I will buy it." Model menunjukkan keyakinan tinggi dalam klasifikasi ini, yang mencerminkan pandangan positif penulis terhadap mobil tersebut.

**Summarization**

"summarization" berhasil merangkum informasi tentang Nabire menjadi bentuk yang lebih singkat dan padat. Model mampu mengekstrak poin-poin utama, seperti lokasi, sumber daya alam, dan potensi pariwisata, tanpa kehilangan esensi informasi.

**Translation**

pipeline "translation" berhasil menerjemahkan kalimat "aku suka minum kopi" dari bahasa Indonesia ke bahasa Inggris, menghasilkan terjemahan yang akurat: "I like drinking coffee."
